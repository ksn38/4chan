```python
df1 = pd.DataFrame({'rkey': ['moo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['ooo', 'bar', 'baz', 'uoo'],
                    'value': [5, 6, 7, 8]})
```

```python
pd.merge(df1, df2, how='outer', on='rkey')
```
